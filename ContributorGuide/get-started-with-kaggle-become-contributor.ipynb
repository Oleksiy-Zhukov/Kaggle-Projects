```python
### This is a simple code solution for Titanic competition, feel free to change it and explore ###
### Author: https://www.kaggle.com/zhukovoleksiy
#------------------------------------------------------------------------------------------------#

### IMPORT
# data processing, CSV file I/O (e.g. pd.read_csv)
import pandas as pd 

# Modeling and Predicting
from sklearn.tree import DecisionTreeClassifier


### READ DATA
train_data = pd.read_csv('/kaggle/input/titanic/train.csv')
test_data = pd.read_csv('/kaggle/input/titanic/test.csv')


###DATA PREPROCESSING
#drop most of the columns, because our model will be very simple
train_data = train_data.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'SibSp', 'Parch'], axis=1)

#doing the same with test_data
test_data = test_data.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'Embarked', 'SibSp', 'Parch'], axis=1)

#transform 'Sex' column values into '1' or '0'
train_data['Sex'] = train_data['Sex'].replace({'male': 0, 'female': 1})
    
#doing the same with the test_data]
test_data['Sex'] = test_data['Sex'].replace({'male': 0, 'female': 1})

#fill null values in Age column:
mean_age = train_data['Age'].mean()
train_data['Age'] = train_data['Age'].fillna(mean_age).astype('int')

#same with test_data:
test_mean_age = test_data['Age'].mean()
test_data['Age'] = test_data['Age'].fillna(test_mean_age).astype('int')

#Spliting train data into X (features) and y (answear):
y = train_data['Survived']
X = train_data.drop(['Survived'], axis=1)
# We don't need to do this with test_data because test_data has no answears, only features.


### TRAINING MODEL
# here you can change our model parameters such as max_depth or random_state
model = DecisionTreeClassifier(max_depth=4)
model.fit(X, y)


### PREDICTIONS
# Prediction of target for test data
preds = model.predict(test_data).astype(int)

# Saving the result into submission file
submission = pd.read_csv('/kaggle/input/titanic/gender_submission.csv')
submission["Survived"] = preds
submission.to_csv('submission.csv', index=False) # Competition rules require that no index number be saved
```
