## <h2 style="text-align: left; background-color:#1192AA; font-family: garamond; color: #EEE8A9; padding: 13px; line-height:0.9;border-radius:2px; margin-bottom: 0em; text-align: center; font-size: 25px; border-radius: 100px 100px;">IMPORT AND LOAD DATA</h2>

```python
# data processing, CSV file I/O (e.g. pd.read_csv)
import pandas as pd

# data viz
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# students performance in exams data
data_stp = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
data_stp['gender'] = data_stp['gender'].replace({'male': 0, 'female': 1})

# suicide rates overview 1985 - 2016
data_sr = pd.read_csv('../input/suicide-rates-overview-1985-to-2016/master.csv')
data_sr['sex'] = data_sr['sex'].replace({'male': 0, 'female': 1})
```

## <h2 style="text-align: left; background-color:#1192AA; font-family: garamond; color: #EEE8A9; padding: 13px; line-height:0.9;border-radius:2px; margin-bottom: 0em; text-align: center; font-size: 25px; border-radius: 100px 100px;">BASIC PLOTS</h2>

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Choose Palette</h2>

```python
my_palette = sns.cubehelix_palette(n_colors = 7, start=.46, rot=-.45, dark = .2, hue=0.95)
sns.palplot(my_palette)
plt.gcf().set_size_inches(13,2)

for idx,values in enumerate(my_palette.as_hex()):
    plt.text(idx-0.375,0, my_palette.as_hex()[idx],{'font': "Courier New", 'size':16, 'weight':'bold','color':'black'}, alpha =0.7)
plt.gcf().set_facecolor('white')

plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Plot</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
color = my_palette[4]

# plot a sequence of first 50 (not sorted) math scores in dataset
plt.plot(data_stp.head(50)["math score"], color = color)

# set title
plt.title('Line Plot 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
color = my_palette[6]

# plot a sequence of first 25 (not sorted) reading scores in dataset with different line type
plt.plot("reading score", 
         data = data_stp.head(25),
         color = color, ls = '--')

# set title
plt.title('Line Plot 2', fontsize = 14, fontweight = 'bold')

# set legend
plt.legend(['Legend'])

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Scatter</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
color = my_palette[4]

# plot a scatter of reading score vs writing score
plt.scatter(data = data_stp.head(100), x = "reading score", y = "writing score", color = color)

# set x and y labels
plt.xlabel('Reading Score', fontsize = 13)
plt.ylabel('Writing Score', fontsize = 13)

# set title
plt.title('Scatter Plot 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set color map
cmap = sns.dark_palette(my_palette[3], as_cmap=True)

# plot a scatter of math score vs reading score with additional variable gender as color
scatter = plt.scatter(data = data_stp.head(100),
            x = "math score",
            y = "reading score",
            c = "gender",
            alpha = 0.7,
            marker = 'o')

scatter.set_cmap(cmap)

# set x and y labels
plt.xlabel('X Label', fontsize = 13)
plt.ylabel('Y Label', fontsize = 13)

# set legend
plt.legend(*scatter.legend_elements(),
           loc = 'upper left',
           title = 'Legend')

# set title
plt.title('Scatter Plot 2', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Bar</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# plot a bar chart using frequency of generation types
plt.bar(data_sr["generation"].value_counts().index,
        data_sr["generation"].value_counts(),
        color = colors)

# set title
plt.title('Bar Plot 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# plot first bar chart using frequency of race/ethnicity types where gender value is female
plt.bar(x = data_stp.loc[data_stp.gender == 1]["race/ethnicity"].value_counts().index,
        height = data_stp.loc[data_stp.gender == 1]["race/ethnicity"].value_counts(),
        label = "Female",
        width = 0.4,
        edgecolor = 'black',
        color = colors[1])

# plot second bar chart using frequency of race/ethnicity types where gender value is male and align it to the side
plt.bar(x = data_stp.loc[data_stp.gender == 0]["race/ethnicity"].value_counts().index,
        height = data_stp.loc[data_stp.gender == 0]["race/ethnicity"].value_counts(),
        label = "Male",
        width = 0.3,
        align = 'edge',
        edgecolor = 'black',
        color = colors[5])

# display legend
plt.legend()

# set title
plt.title('Bar Plot 2', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# plot a horizontal bar chart using freaquency of generation types
plt.barh(data_sr["generation"].value_counts().index,
        data_sr["generation"].value_counts(),
        color = colors)

# set title
plt.title('Bar Plot 3', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Stem</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# create a stem plot using first 75 values (not sorted) of writing score, set a bottom line to 60
stemplot = plt.stem(data_stp["writing score"].head(75),
         markerfmt='D',
         bottom=60)

# set line color
stemplot.stemlines.set_color(colors[1])

# set markerline color
stemplot.markerline.set_markerfacecolor(colors[3])

# set baseline (bottom line) color
stemplot.baseline.set_color(colors[5])

# set title
plt.title('Stem Plot 1', fontsize = 14, fontweight = 'bold')

# set x and y labels
plt.ylabel('Writing Score')
plt.xlabel('Student')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette


# create a stem plot using last 25 values (not sorted) of population values
stemplot = plt.stem(data_sr["population"].tail(25),
         markerfmt='o')

# set line color
stemplot.stemlines.set_color(colors[6])

# set markerline color
stemplot.markerline.set_markerfacecolor(colors[0])

# set baseline (bottom line) color
stemplot.baseline.set_color(colors[5])


# set title
plt.title('Stem Plot 2', fontsize = 14, fontweight = 'bold')

# set x and y labels
plt.ylabel('Population')
plt.xlabel('')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Step</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# plot a step chart using reading score of last 30 students in dataset (not sorted)
plt.step(x = data_stp["reading score"].tail(30).index,
        y = data_stp["reading score"].tail(30), where='post', color = my_palette[4])

# set title
plt.title('Step Plot 1', fontsize = 14, fontweight = 'bold')

# set x and y labels
plt.ylabel('Reading Score')
plt.xlabel('Student')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# plot a step chart using writing score of last 30 students in dataset (not sorted)
plt.step(x = data_stp["writing score"].tail(30).index,
        y = data_stp["writing score"].tail(30), where='post', color = my_palette[1])

# add a line plot of the same values for styling
plt.plot(data_stp["writing score"].tail(30), 'o-', color=my_palette[5], alpha=0.5)

# set title
plt.title('Step Plot 2', fontsize = 14, fontweight = 'bold')

# set x and y labels
plt.ylabel('Writing Score')
plt.xlabel('Student')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Fill Between</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# plot a fill between chart using writing score of first 25 students (not sorted) in dataset
plt.fill_between(x = data_stp["writing score"].head(25).index,
                 y1 = data_stp["writing score"].head(25),
                 y2 = data_stp["reading score"].head(25),
                 color = my_palette[2], alpha = 0.7, edgecolor = my_palette[4], linewidth = 1.5)

# add a line plot of the same values for styling
plt.plot(data_stp["writing score"].head(25), 'D-', color=my_palette[5])

# set title
plt.title('Fill Between Plot 1', fontsize = 14, fontweight = 'bold')

# set x and y labels
plt.ylabel('Writing Score')
plt.xlabel('Student')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Stack</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette.reverse()

# select features (first 25 values of every exam score)
features = [data_stp['math score'].head(25), data_stp['reading score'].head(25), data_stp['writing score'].head(25)]

# set labels
labels = ["math score", "reading score", "writing score"]

# create a stack plot using our features
plt.stackplot(data_stp.head(25).index, features,
              colors = my_palette[3:], edgecolor = 'b', alpha = 0.8)

# set legend with labels
plt.legend(labels)

# set title
plt.title('Stackplot 1', fontsize = 14, fontweight = 'bold')

# set x and y labels
plt.ylabel('Exam Score')
plt.xlabel('Student')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Hist</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

plt.hist(data_stp["reading score"],
        color = my_palette[1], edgecolor = 'k')
plt.title('Hist Plot 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

plt.hist(data_stp["writing score"],
         color = my_palette[2], bins = 9,
         edgecolor = my_palette[2],
         cumulative = True,
         histtype = 'stepfilled',
         density=True,
         alpha = 0.9
        )

plt.xticks(range(0, 101, 10))
plt.title('Hist Plot 2', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

plt.hist(data_stp["math score"],
        color = my_palette[4],
        bins = 20, alpha = 0.8, edgecolor = my_palette[0]
        )
plt.plot(data_stp["math score"].value_counts().sort_index(), '--',
        color = my_palette[1], marker='o')

plt.xticks(range(0, 101, 5))
plt.title('Hist Plot 3', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Boxplot</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

plt.boxplot(data_stp['reading score'], sym='', widths=0.3, 
            capprops = dict(linestyle='--', linewidth=2, color=my_palette[4]),
            boxprops = dict(linestyle='-', linewidth=2, color=my_palette[2]),
            whiskerprops = dict(linestyle='-', linewidth=2, color=my_palette[2]))


plt.title('Boxplot 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette[4:]

# set labels
labels = ['ex1', 'ex2', 'ex3']

bplot = plt.boxplot([data_stp['reading score'], data_stp['math score'], data_stp['reading score']],
                    notch = True,  sym='', widths=0.3, labels = labels,
                    patch_artist=True, 
                    capprops = dict(linestyle='--', linewidth=2, color=my_palette[2]),
                    boxprops = dict(linestyle='-', linewidth=2, color=my_palette[2]),
                    whiskerprops = dict(linestyle='-', linewidth=2, color=my_palette[2]))

# fill with colors
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Boxplot 2', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Errorbar</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

plt.errorbar(data_sr["generation"].value_counts().index,
             data_sr["generation"].value_counts(), yerr = 1000, 
             fmt = 'o--',
             ecolor = my_palette[4],
             color = my_palette[1],
             
        )

plt.title('Errorbar 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Violinplot</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# set labels
labels = ['ex1', 'ex2', 'ex3']

vlp = plt.violinplot([data_stp['reading score'], data_stp['math score'], data_stp['reading score']],)
            
for pc in vlp['bodies']:
    pc.set_facecolor(colors[4])
    pc.set_edgecolor(colors[2])
    pc.set_alpha(0.6)
    
vlp['cmins'].set_color(colors[1])
vlp['cmaxes'].set_color(colors[1])
vlp['cbars'].set_color(colors[1])
    
plt.title('Violinplot 1', fontsize = 14, fontweight = 'bold')

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

## <h2 style = "font-family:garamond; font-size:30px; background-color: white; color : #1192AA; border-radius: 100px 100px; text-align:left">Pie</h2>

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# set labels
labels = 'Lable 1', 'Lable 2'

plt.pie(data_stp['lunch'].value_counts(),
        labels = labels,
        colors = colors[2:],
        autopct='%1.1f%%',
        wedgeprops = dict(width=1, edgecolor=colors[6]),
        textprops = dict(fontsize = '13'))

plt.title('Pie 1', fontsize = 14, fontweight = 'bold')
plt.legend()

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```

```python
# create figure and set style with white background
plt.figure(figsize = (14, 8))
sns.set_style('white')

# set colors
colors = my_palette

# set labels
labels = ['L 1', 'L 2', 'L 3', 'L 4', 'L 5']

plt.pie(data_stp['race/ethnicity'].value_counts(),
        explode = (0.1, 0, 0, 0, 0),
        shadow=True,
        colors = colors[2:],
        labels = labels,
        autopct='%1.1f%%',
        wedgeprops = dict(width=1, edgecolor='k'),
        textprops = dict(fontsize = '13'))

plt.title('Pie 2', fontsize = 14, fontweight = 'bold')
plt.legend()

# remove spines from plot
sns.despine()

# display all open figures
plt.show()
```
